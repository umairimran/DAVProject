```python
import os
import pandas as pd

def combine_csv_files(folder_path):
    """
    Reads all CSV files in a folder, combines them into a single DataFrame, and returns the DataFrame.

    :param folder_path: Path to the folder containing the CSV files.
    :return: A combined DataFrame with all the CSV data.
    """
    # List to store each DataFrame
    dataframes = []

    # Loop through each file in the folder
    for file_name in os.listdir(folder_path):
        # Check if the file is a CSV
        if file_name.endswith(".csv"):
            file_path = os.path.join(folder_path, file_name)
            
            # Read the CSV into a DataFrame
            df = pd.read_csv(file_path)
            dataframes.append(df)  # Add the DataFrame to the list
            print(f"Loaded {file_name}")

    # Combine all DataFrames into a single DataFrame
    combined_df = pd.concat(dataframes, ignore_index=True)
    return combined_df

# Example usage
folder_path = "E:/worldBankCsv"  # Replace with your folder path containing CSV files
combined_df = combine_csv_files(folder_path)
print(combined_df)
```

```python
import pandas as pd
import glob
import os

# Define the folder path where your CSV files are stored
folder_path = "E://worldBankCsv//"  # Update this to your folder path

# Get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Define the column names as per your previous transformation
column_names = [
    'Country Name', 'Country Code', 'Indicator Name', 'Indicator Code',
     '1963', '1964', '1965', '1966', '1967', '1968',
    '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981',
    '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994',
    '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007',
    '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
    '2021', '2022', '2023','2024'
]

# Initialize an empty list to hold the DataFrames
df_list = []

# Loop through each CSV file and apply the transformations
for file in csv_files:
    # Read the CSV file, skipping the first few rows and handling encoding
    df = pd.read_csv(file, encoding='utf-8', skiprows=[0, 1, 3, 4])
    
    # Exclude the last 3 columns as per your requirement
    df_excluded = df.iloc[:, :-3]
    df_excluded=pd.DataFrame(df_excluded.index)
    # Convert the first column (index) into a list of values
    temp_df = pd.DataFrame(df_excluded[0].apply(lambda x: list(x) if isinstance(x, tuple) else x).tolist(), columns=column_names)
    
    # Append the transformed DataFrame to the list
    df_list.append(temp_df)

# Concatenate all DataFrames into one single DataFrame, maintaining consistency
final_df = pd.concat(df_list, ignore_index=True)

# Display the first few rows of the final DataFrame to verify
print(final_df.head())
```

```python
# Assuming 'final_df' is your DataFrame
# Select columns from '2000' to '2023' onwards (keep those)
columns_to_keep = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'] + [str(year) for year in range(2000, 2024)]
df = df[columns_to_keep]

```

```python
import pandas as pd

# Step 1: Melt the DataFrame to long format, keeping country and indicator as ID variables
melted_df = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                          var_name='Year', 
                          value_name='Value')

# Step 2: Convert Year column to integer type
melted_df['Year'] = melted_df['Year'].astype(int)

# Step 3: Create a pivot table where 'Country Name' and 'Year' are the indices
pivot_df = melted_df.pivot_table(index=['Country Name', 'Year'], 
                                 columns='Indicator Name', 
                                 values='Value', 
                                 aggfunc='first')

# Step 4: Reset the index to flatten the DataFrame
pivot_df.reset_index(inplace=True)

# Display the resulting DataFrame
print(pivot_df.head())
```

```python
import pandas as pd

# Load the two CSV files into pandas DataFrames
df1 = pd.read_csv('C:\\Users\\hp\\Desktop\\health_care_ML_project\\models\\data.csv')
df2 = pivot_df  # Assuming pivot_df is already defined as your second DataFrame

# Rename columns if necessary (ensure both have 'Country Name' and 'Year')
df1.rename(columns={'name': 'Country Name', 'year': 'Year'}, inplace=True)
df2['Year Disclosed']=df2['Year']
# Clean up the 'Country Name' column in both DataFrames by removing leading/trailing spaces
df1['Country Name'] = df1['Country Name'].str.strip()
df2['Country Name'] = df2['Country Name'].str.strip()

# Ensure 'Year' columns are of the same type (string)
df1['Year'] = df1['Year'].astype(str)
df2['Year'] = df2['Year'].astype(str)

# Perform an inner join on both 'Country Name' and 'Year' columns
merged_df = pd.merge(df1, df2, on=['Country Name', 'Year'], how='inner', suffixes=('_df1', '_df2'))

# Display the merged DataFrame
print(merged_df.head())
```
